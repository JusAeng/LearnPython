```python
import pandas as pd
from matplotlib import pyplot as plt
import random
```

```python
data=pd.read_csv("train.csv")
data
```

```python
age=data['experience']
salary=data['salary']
salary.head()
# plt.plot_date(age,salary)
```

```python
rand_salarys=[]
for i in range(len(salary)):
    temp=[int(x) for x in salary[i].split("to")]
    est_salary=random.uniform(temp[0],temp[1])
    rand_salarys.append(est_salary)
data["est_salary"]=rand_salarys 
```

```python
rand_age=[]
for i in range(len(age)):
    use,cache=[x for x in age[i].split()]
    temp=[int(x) for x in use.split("-")]
    est_age=random.uniform(temp[0],temp[1])
    rand_age.append(est_age)
data["est_age"]=rand_age
data.head()
```

```python
location=data["location"]
all_Location=[]
temp_location=[]
for i in range(len(location)):
    temp_location=[x.strip() for x in location[i].split(",")]
    for ele in temp_location:
        if ele not in all_Location:
            all_Location.append(ele)
print(len(all_Location))

for j in all_Location:
    print(j)
```
